```python
print("===================================")
print("      Welcome to Basic Chatbot")
print("===================================")
print("Type 'bye' to exit the chatbot.\n")

while True:

    user = input("You: ").lower().strip()

    if user == "hello" or user == "hi":
        print("Bot: Hello! Nice to meet you.")

    elif user == "how are you":
        print("Bot: I'm fine, thanks! How are you?")

    elif user == "i am fine" or user == "i'm fine":
        print("Bot: That's great to hear!")

    elif user == "what is your name":
        print("Bot: I'm a Basic Python Chatbot.")

    elif user == "who created you":
        print("Bot: I was created by Siddhant using Python.")

    elif user == "help":
        print("Bot: You can ask me:")
        print("- hello")
        print("- how are you")
        print("- what is your name")
        print("- who created you")
        print("- bye")

    elif user == "thank you" or user == "thanks":
        print("Bot: You're welcome!")

    elif user == "bye":
        print("Bot: Goodbye! Have a great day.")
        break

    else:
        print("Bot: Sorry, I don't understand that.")
```
